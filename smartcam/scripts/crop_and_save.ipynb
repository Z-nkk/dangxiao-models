```python
# crop images and save the cropped ones
import cv2
from matplotlib import pyplot as plt
```

```python
img = cv2.imread("../data/classroom/to_label/5107_1_yuanmu_test1_1637206994_1020.jpg")
img2 = img[:,:,::-1]
plt.imshow(img2)
```

```python
x = 600
w = 250
y = 300
h = 300

crop_img = img[y:y+h, x:x+w]
crop_img2 = crop_img[:,:,::-1]
plt.imshow(crop_img2)
```

```python
x = 920
w = 250
y = 280
h = 300

crop_img = img[y:y+h, x:x+w]
crop_img2 = crop_img[:,:,::-1]
plt.imshow(crop_img2)
```

```python
import os

img_dir = "../data/classroom/to_label"
out_dir = "../data/classroom/to_label_cropped"

for filename in os.listdir(img_dir):
    if filename.endswith(".jpg"):
        img = cv2.imread(f"{img_dir}/{filename}")
        
        x = 600
        w = 250
        y = 300
        h = 300
        crop_img = img[y:y+h, x:x+w]
        cv2.imwrite(f"{out_dir}/{filename[:-4]}_zyf.jpg", crop_img)
        
        x = 920
        w = 250
        y = 280
        h = 300
        crop_img = img[y:y+h, x:x+w]
        cv2.imwrite(f"{out_dir}/{filename[:-4]}_ljy.jpg", crop_img)
```
